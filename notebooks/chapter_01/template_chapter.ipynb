## 2. Examples from the Book

### Example 1: [Location Estimates of population and murder]
Description...

```python
# Example code

import pandas as pd
from scipy.stats import trim_mean

# Load data
df_state = pd.read_csv('../../data/raw/state.csv')

# Calculate location estimates
mean = df_state['Population'].mean()
trimmed_mean = trim_mean(df_state['Population'], 0.1)
median = df_state['Population'].median()

print(f"The mean population is {mean}.")
print(f"The trimmed mean population is {trimmed_mean}.")
print(f"The median population is {median}.")


```

```python
import numpy as np
import wquantiles

weighted_mean = np.average(df_state['Murder.Rate'], weights=df_state['Population'])
print(f"The weighted mean murder rate is {weighted_mean}.")

weighted_quantile = wquantiles.median(df_state['Murder.Rate'], df_state['Population'])
print(f"The weighted median murder rate is {weighted_quantile}.")









```

```python
from statsmodels import robust

standard_deviation = df_state['Population'].std()
print(f"The standard deviation of the population is {standard_deviation}.")

quantile_deviation = df_state['Population'].quantile(0.75) - df_state['Population'].quantile(0.25)
print(f"The quantile deviation of the population is {quantile_deviation}.")

robust_deviation = robust.scale.mad(df_state['Population'])
print(f"The robust deviation of the population is {robust_deviation}.")





```

```python
# Percentiles and Boxplots
df_state[["Murder.Rate"]].quantile([0.05, 0.25, 0.5, 0.75, 0.95])





```

```python
ax = (df_state["Population"] / 1_000_000).plot.box(figsize=(3, 4))
ax.set_ylabel("Population (millions)")
```

```python

binned_population = pd.cut(df_state["Population"], 10)
# Create a DataFrame to display bin number, bin range, count, and state names
bin_counts = binned_population.value_counts().sort_index()
bin_edges = binned_population.cat.categories

result = []
for idx, (interval, count) in enumerate(zip(bin_edges, bin_counts)):
    states_in_bin = df_state[binned_population == interval]["State"].tolist()
    result.append({
        "Bin Number": idx + 1,
        "Bin Range": str(interval),
        "Count": count,
        "States": ", ".join(states_in_bin)
    })

import pandas as pd
pd.DataFrame(result)
```
